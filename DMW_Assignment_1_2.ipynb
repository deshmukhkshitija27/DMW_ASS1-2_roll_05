```python
data=pd.read_csv('googleplaystore_v2.csv')
```

```python
#cleaning
data
```

```python
data.isnull().sum()
```

```python
data1=data.fillna(method="ffill")
```

```python
data1.isnull().sum()
```

```python
data1
```

```python
#transformation
data1.head()
```

```python
dummy=pd.get_dummies(data1['Type'])
```

```python
dummy
```

```python
data2=pd.concat((data1,dummy),axis=1)
```

```python
data2=data2.drop(['Type'],axis=1)
```

```python
data2.head()
```

```python
data2=data2.drop(['Free'],axis=1)
```

```python
data2=data2.rename(columns={'Paid':'Type'})
```

```python
data2.head()
```

```python
#reduction
data2=data2.drop(['Reviews'],axis=1)
```
